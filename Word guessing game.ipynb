## Function to fetch the words from the wordslist file

```python
import random

def get_random_word():
    random_word = None
    with open('wordslist.txt','r') as file:
        for words in file:
            words = words.strip().lower()
            word = words.split(',')
    random_word = random.choice(word)
    return random_word
# get_random_word()
```

## Simply assign to a variable instead of function

```python
import random

veggies_list = ['cauliflower','cabbage','capsicum','bittergourd','carrot','pumpkin','zucchini','frenchbean','okra','greenpea','spinach','lettuce','corn','brussels','onion','turnip','potato','beetroot','radish','broccoli']
random_word = random.choice(veggies_list)
# print(random_word)

# Here the user is asked to enter their name
name = input("Enter your name: ") 
print("Welcome", name,"!")
print("Let's play! Guess the vegetable name")
guesses = '' # input letter stored here
attempts = 20 # initiate any no of attempts
while attempts > 0:
    failed = 0
    for letter in random_word:
        if letter in guesses:
            print(letter)
        else:
            print("*")
            failed += 1
    if failed == 0:
        print("Hurray!!!")
        print("The vegetable name is:",random_word)
        print("Congrats!",name,"You won the game")
        break
        
    guess = input("Guess a letter:") 
    guesses += guess
    if guess not in random_word:
        attempts -= 1
        print("Incorrect")
        print("You have", + attempts, 'more guesses')
        if attempts == 0: 
            print("Sorry!",name,"You Loose the game")    
    
```
